```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
Tbs= lb.fit_transform(df[['Total Compensation and Benefits']])
```

```python
Tbs
```

```python
# Convert columns to string type before applying string methods
df['Total Compensation and Benefits'] = df['Total Compensation and Benefits'].astype(str)
df['Bonuses'] = df['Bonuses'].astype(str)
df['Overtime Pay'] = df['Overtime Pay'].astype(str)
df['Incentives'] = df['Incentives'].astype(str)
df['Base Salary'] = df['Base Salary'].astype(str)

# Now you can apply string methods
df['Total Compensation and Benefits'] = df['Total Compensation and Benefits'].str.replace("$","").str.replace(",","").astype(float)
df['Bonuses'] = df['Bonuses'].str.replace("$","").str.replace(",","").astype(float)
df['Overtime Pay'] = df['Overtime Pay'].str.replace("$","").str.replace(",","").astype(float)
df['Incentives'] = df['Incentives'].str.replace("$","").str.replace(",","").astype(float)
df['Base Salary'] = df['Base Salary'].str.replace("$","").str.replace(",","").astype(float)
```

```python
df['Salary'] = df['Salary'].str.replace("$","").str.replace(",","")
```

```python
X=Tbs.reshape(-1, 1)
z=df['Overtime Pay']
b=df['Base Salary']
```

```python
y=df['Salary'].str.replace("$","").str.replace(",","")
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainx,testx,trainy,testy = train_test_split(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mod = LinearRegression()
```

```python
mod.fit(trainx,trainy)
```

```python
df['Pred_sal']=mod.predict(X)
```
